```python
def get_places(text):
    
      
    place_freq = collections.defaultdict(int)
    for entity in text.ents:
        if (entity.label_ == 'GPE') or (entity.label_ == 'LOC'):
            place_freq[entity.text] += 1 # the utility of defaultdict!
    place_freq = dict(place_freq)

    #to df
    place_freq_df = pd.DataFrame.from_dict(place_freq, orient='index').reset_index().rename(columns={'index':'place_name',0:'frequency'})

    return place_freq_df


places = get_places(processed_lyrics)
places_sorted = places.sort_values('frequency', ascending = False)



places_sorted.to_csv('bb2.csv')
```

```python
# importing geopy library

from geopy.geocoders import Nominatim
import time
 
# calling the Nominatim tool
loc = Nominatim(user_agent="GetLoc")
 


lat = []
long = []

def get_coords(places_test):

    for i in places_test['place_name']:
        try:
            getLoc = loc.geocode(i)
            print(f'{i}s address is')
            print(getLoc.address)
            print("Latitude = ", getLoc.latitude,)
            lat.append(getLoc.latitude)
            print("Longitude = ", getLoc.longitude, "\n")
            long.append(getLoc.longitude)
            
        except:
            lat.append(None)
            long.append(None)
            time.sleep(1)



    places_test['Lat'] = lat
    places_test['Long'] = long

    return places_test



bb2 = get_coords(places_test)
```
