```python
## function to get playlist tracks

references = 'spotify:playlist:0CHwZd4tvPNqY71x10fN1r'
where_u_r = 'spotify:playlist:6Cmr2W6FfsfbOpdv7lIeVw'
role_model = 'spotify:playlist:5fsF3jPI7Hh3Ojel1J0DYA'
bb_small = 'spotify:playlist:2gi9JkknI6JPatl37nIQZp'
my_hair = 'spotify:playlist:0v3fcb0DxHz6JGLJSCpGd4'

def get_playlist_tracks(uri_info):
    uri = []
    track = []
    artist = []
    duration = []
    explicit = []
    track_number = []

    one = sp.playlist_items(uri_info, limit=50, offset=0, market='US')
    df1 = pd.DataFrame(one)



    for i, x in df1['items'].items():
        uri.append(x['track']['uri'])
        track.append(x['track']['name'])
        artist.append(x['track']['artists'][0]['name'])
        duration.append(x['track']['duration_ms'])
        explicit.append(x['track']['explicit'])
        track_number.append(x['track']['track_number'])
    
    df2 = pd.DataFrame({
        'uri':uri,
        'track':track,
        'artist':artist,
        'duration_ms':duration,
        'explicit':explicit,
        'track_number':track_number})
    
    return df2

   


playlist = get_playlist_tracks(my_hair)
playlist
```

```python


genius = lyricsgenius.Genius("Y2SaEDNEyen2GkoALWF_xF54f5E6X71HIkhrd2QtCXxJqgqy_oSN8cwwd043Kg0a")

def scrape_lyrics(artist, title):

    num = title.find('feat.')
    if num == -1:
        num = title.find('ft.')

    if num == -1:
        num = title.find(' - ')
    
    if num != -1:
        title = title[:num]

    
        

    song = genius.search_song(title, artist)

    if song == None:
        return ""


    lyrics = genius.lyrics(song.id, remove_section_headers=True)

    if ((song.artist == 'Spotify') or (song.artist == 'spotify')):
        return ""

    end = lyrics.find('Embed')
    beg = lyrics.find('Lyrics') + 6

    
    

    subbed_lyrics = lyrics[beg:end]

    return subbed_lyrics


playlist_lyrics = ''

for ind in playlist.index:
    text = scrape_lyrics(playlist['artist'][ind], playlist['track'][ind])
    playlist_lyrics = playlist_lyrics + '\n' + text


```

```python
#process text
#don't care to deal with non-lyric entities, because doesn't matter for geoparsing :)





nlp = spacy.load("en_core_web_sm")

processed_lyrics = nlp(playlist_lyrics)

processed_lyrics

# tokens = [token for token in processed_lyrics]
# pnoun_tokens = [token for token in tokens if token.pos_ == 'PROPN'] # insert list comprehension here
# pnoun_lemmas = [token.lemma_ for token in pnoun_tokens] # insert list comprehension here


# print(pnoun_lemmas)
```

```python
def get_places(text):
    
      
    place_freq = collections.defaultdict(int)
    for entity in text.ents:
        if (entity.label_ == 'GPE') or (entity.label_ == 'LOC'):
            place_freq[entity.text] += 1 # the utility of defaultdict!
    place_freq = dict(place_freq)

    #to df
    place_freq_df = pd.DataFrame.from_dict(place_freq, orient='index').reset_index().rename(columns={'index':'place_name',0:'frequency'})

    return place_freq_df


places = get_places(processed_lyrics)
places_sorted = places.sort_values('frequency', ascending = False)

places_sorted.to_csv('output.txt')
```
